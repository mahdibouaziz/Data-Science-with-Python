## Concat 2 DataFrames 

* Create a DataFrame

```python
df1=pd.DataFrame ({"Name":["mahdi","ahmed","fatma","sami"],
                 "Age":[20,19,21,25],
                 "Stars":[3,5,4,4]})
df1
```

```python
df2=pd.DataFrame({"Name":["hadir","oussema"],
                 "Age":[20,23],
                 "Stars":[3,4]})
df2
```

```python
#Concat rows
df4=pd.concat([df1,df2],ignore_index=True)
df4
```

```python
df3=pd.DataFrame ({"Name":["mahdi","fatma","hadir"],
                  "klg":[72,65,55]})
```

```python
#Concat columns
pd.concat([df4,df3],axis=1,ignore_index=True)
#as we see it will reset columns name
```

```python
#You can merge on one column
pd.merge(df4,df3, on="Name")
#it take the intersection
```

```python
#if you want to take the union just put how="outer"
pd.merge(df4,df3, on="Name",how="outer")
```
